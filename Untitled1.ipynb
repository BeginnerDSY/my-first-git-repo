```python
import pandas as pd
import os

data_path = os.path.expanduser('~/autodl-tmp/ptbxl_data/ptb-xl-a-large-publicly-available-electrocardiography-dataset-1.0.3/ptbxl_database.csv')

df = pd.read_csv(data_path)

print(df.columns.tolist())
print(df[['ecg_id', 'filename_hr', 'filename_lr']].head())
```

```python
import os
import wfdb
import matplotlib.pyplot as plt

# PTB-XL 根目录
root_dir = os.path.expanduser('~/autodl-tmp/ptbxl_data/ptb-xl-a-large-publicly-available-electrocardiography-dataset-1.0.3')

# 取第一条记录
row = df.iloc[0]

# 拼成完整路径（注意不要加 .dat/.hea 后缀）
record_path = os.path.join(root_dir, row['filename_hr'])

print("ecg_id:", row['ecg_id'])
print("record_path:", record_path)

# 读取波形
signal, fields = wfdb.rdsamp(record_path)

print("signal.shape =", signal.shape)
print("sig_name =", fields['sig_name'])

# 画一个导联，先看 II 导联
lead_names = fields['sig_name']
lead_idx = lead_names.index('II') if 'II' in lead_names else 0

plt.figure(figsize=(12, 3))
plt.plot(signal[:, lead_idx])
plt.title(f"ECG ID {row['ecg_id']} - Lead {lead_names[lead_idx]}")
plt.xlabel("sample")
plt.ylabel("amplitude")
plt.show()
```

```python
import pandas as pd
import os
import ast

root_dir = os.path.expanduser('~/autodl-tmp/ptbxl_data/ptb-xl-a-large-publicly-available-electrocardiography-dataset-1.0.3')
scp_map = pd.read_csv(os.path.join(root_dir, 'scp_statements.csv'), index_col=0)

row = df.iloc[0]
codes = ast.literal_eval(row["scp_codes"])

print("原始scp_codes:", codes)
print("\n翻译后：")
for code, likelihood in codes.items():
    if code in scp_map.index:
        print(code, "|", "likelihood =", likelihood, "|", scp_map.loc[code].to_dict())
    else:
        print(code, "| likelihood =", likelihood, "| 未在scp_statements.csv中找到")
```

```python
import ast

# 1. 提取我们刚才画图的那位患者（第一行，row）的诊断码
scp_codes_str = row['scp_codes']
print(f"原始诊断记录: {scp_codes_str}")

# 2. 把字符串变成真正的字典，方便我们提取
scp_codes_dict = ast.literal_eval(scp_codes_str)

# 3. 把所有的疾病缩写（字典的键）单独拎出来
diagnoses = list(scp_codes_dict.keys())
print(f"最终确诊标签: {diagnoses}")
```
